```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. loading data
df = pd.read_csv('./data/college_completion.csv', low_memory=False)
```

```python
# 2. dimensions of data, observations, variables
print(df.shape, '\n')   # 3798 observations, 63 variables
print(df.head(), '\n')
```

```python
# 3. cross tabulate
print(pd.crosstab(df['control'], df['level']), '\n')
print(pd.crosstab(df['control'], df['level'], normalize='all'), '\n')

# Patterns:
# The most common institutions are private not-for-profit 4 year, 
# next most common is public 2 year institutions, then public 4 year 
# and private for-profit 2/4 years. Most 4 year institutions are public.
```

```python
# 4. create histogram
import seaborn as sns
sns.histplot(df['grad_100_value'])
```

```python
# 4. kernel desity plot
sns.kdeplot(df['grad_100_value'])
```

```python
# 4. boxplot
sns.boxplot(df['grad_100_value'])
```

```python
# 4. statistical description
df['grad_100_value'].describe()
```

```python
# 5. grouped kernel desntiy plot, control
sns.kdeplot(data=df,x='grad_100_value',hue='control')
```

```python
# 5. grouped kernel desntiy plot, level
sns.kdeplot(data=df,x='grad_100_value',hue='level')
```

```python
# 5. make grouped calculations
df.loc[:,['control','grad_100_value']].groupby('control').describe()
```

```python
# 5. make grouped calculations
df.loc[:,['level','grad_100_value']].groupby('level').describe()
```

4 year average graudation rates are 10 points higher than 2 year, and private not-for-profit rates are around 42%, compared to 29% for private for-profit and 17% for public. 

```python
# 6. grouped kernel density plot
df['levelXcontrol']=df['level'] + ', ' + df['control']
sns.kdeplot(data=df,x='grad_100_value',hue='levelXcontrol')
```

```python
# 6
df.loc[:,['level', 'control', 'grad_100_value']].groupby(['level', 'control']).describe()
```

The institutions that have the best graduation rate is the private for-profit 2 year which is double private for-profit 4 year rate. There's a little difference between 2 year and 4 year programs from private not-for-profit, but public 2-year rate are the worst.

```python
# 7. kernel density plot of aid_value, level
import numpy as np
df['aid_arcsinh'] = np.arcsinh(df['aid_value'])
sns.kdeplot(data=df,x='aid_arcsinh')
```

```python
sns.kdeplot(data=df,x='aid_value')
```

```python
# 7. kernel density plot of aid_value, control
sns.kdeplot(data=df,x='aid_value',hue='control')
```

```python
sns.kdeplot(data=df,x='aid_value',hue='level')
```

```python
df.loc[:,['control','aid_value']].groupby('control').describe()
```

```python
df.loc[:,['level','aid_value']].groupby('level').describe()
```

The second mode is for private not-for-profit institutions, the peak is around 9 which corresponds to public and private for-profit institutions.

```python
# 8. scatterplot of grad_100_value by aid_value
sns.scatterplot(data=df, x='aid_value', y='grad_100_value')
```

There's a group of points of positive association between the two variables. Another group where grad_100_value grows but aid_value doesn't grow much.

```python
sns.scatterplot(data=df,x='aid_value',y='grad_100_value',hue='levelXcontrol')
```

The positive association appears to be mainly 4 year, private NFP institutions. All other groups are not as strong of an association (with not much of a strong direction, not strong positive).

```python
# use of chat here to find missing columns
# Identify columns with missing values
missing_columns = df_4.columns[df_4.isnull().any()]

# Display the results
if not missing_columns.empty:
    print("Columns with missing values:")
    for column in missing_columns:
        print(f"- {column}")
else:
    print("No missing values found in any columns.")
```

Data dictionary document notes that due to the new skip patterns in the Facility Instrument and inclusion of administrative MDS variables when a CCN is reported, values for some variables may be coded as Missing rather than Not indicated. Data users should take this into account when calculating estimates or comparing distributions with prior years. There seems to be no difference between categorical and quantitative.

```python
import pandas as pd

# help of chatgpt
# selection of a few variables
selected_vars = ['ADM_H_MEDSTA', 'ADM_H_GHPSW', 'ADM_OP_MDCD'] + \
                [f'PUFF{str(i).zfill(3)}' for i in range(91, 101)]
# checking for missing values in selected variables
missing_data = df_4[selected_vars].isnull().sum()
print("Missing values in selected variables:\n", missing_data)
# drop rows with missing values
data_cleaned = df_4.dropna(subset=selected_vars)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# help of chatgpt
# boxplots for numeric variables (PUFF091 to PUFF100)
for var in selected_vars[3:]:  # Skip the first three categorical variables
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=data_cleaned[var])
    plt.title(f'Boxplot of {var}')
    plt.show()

# bar graphs for categorical variables
for var in selected_vars[:3]:  # first three categorical variables
    plt.figure(figsize=(10, 6))
    sns.countplot(x=data_cleaned[var])
    plt.title(f'Bar graph of {var}')
    plt.show()
```

```python
# help of chatgpt to generate kernel density plot
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['PUFF091'], bins=30, kde=True)
plt.title('Histogram and Kernel Density Plot of PUFF091')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_cleaned['PUFF091'], y=data_cleaned['PUFF092'])
plt.title('Scatter Plot of PUFF091 vs PUFF092')
plt.xlabel('PUFF091')
plt.ylabel('PUFF092')
plt.show()
```
